```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn

from bokeh.io import output_notebook
from bokeh.io import show
from bokeh.plotting import figure
from bokeh.io import show
from bokeh.layouts import column
from bokeh.models import ColumnDataSource, RangeTool


output_notebook()

plt.rcParams.update({'figure.max_open_warning': 0})
%matplotlib inline 

df = pd.read_csv('Air_Quality.csv',
                dtype={
                    'LocationNamme': 'category',
                    'MeasurementType': 'category',
                    'MeasurementUnit': 'category',
                    'location': 'category'
                })

#df.set_index('ThingID')

df['ObservationTime'] = pd.to_datetime(df['ObservationTime'], infer_datetime_format=True)

df = df[df.ThingID != 66]  # ------- 
df = df[df.ThingID != 84]  #
df = df[df.ThingID != 90]  # These are busted devices reporting weird values.
df = df[df.ThingID != 96]  # 
df = df[df.ThingID != 102] # ------- 

df = df.loc[df.MeasurementValue != -99.99] # removing the bads
```

```python
df.info()
```

```python
df.groupby('ThingID')['ObservationID'].nunique()
```

```python
myTable = df.pivot_table(index=["ThingID", "ObservationTime"], 
                         columns='MeasurementType', 
                         values='MeasurementValue', 
                         fill_value=0, dropna=True)
```

```python
myTable.info()
```

```python
myTable
```

```python
import matplotlib.pyplot as pp
pp.rcParams['figure.figsize'] = [15, 5]

def time_plot(thingid, measurement):
    data = myTable.loc[thingid, measurement]
    pp.plot(data.index, data.values, color='darkorange')
```

```python
time_plot(120, 'Temperature')
```

```python
time_plot(120, 'PM2.5 Particulates')
```

```python
time_plot(120, 'Humidity')
```
