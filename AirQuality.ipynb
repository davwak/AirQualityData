```python
import pandas as pd
import numpy as np
import matplotlib as plt

from bokeh.io import output_notebook
from bokeh.io import show
from bokeh.plotting import figure

output_notebook()

plt.rcParams.update({'figure.max_open_warning': 0})
%matplotlib inline 

df = pd.read_csv('Air_Quality.csv',
                dtype={
                    'LocationNamme': 'category',
                    'MeasurementType': 'category',
                    'MeasurementUnit': 'category',
                    'location': 'category'
                })

#df.set_index('ThingID')

df['ObservationTime'] = pd.to_datetime(df['ObservationTime'], infer_datetime_format=True)

df = df[df.ThingID != 84]  # ------
df = df[df.ThingID != 102] # These are busted devices reporting weird values.
df = df[df.ThingID != 90]  # nuking them
df = df[df.ThingID != 96]  #------- 

df = df.loc[df.MeasurementValue != -99.99] # removing the bads
```
